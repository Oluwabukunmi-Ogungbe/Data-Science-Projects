```python
filepath ='C:/allAccountsSending/github projects/loan_data.csv'
loan_data = pd.read_csv(filepath)
loan_data
```

```python
loan_data = loan_data.astype({'purpose':'string'})

```

```python
loan_data.iloc[:,-1].value_counts()

```

```python
# encode categorical variables to numeric using the ordinal encoder
feature = 'purpose'
encode = LabelEncoder()
loan_data[feature] = encode.fit_transform(loan_data[feature])
loan_data



```

```python
input_data = loan_data.iloc[:,0:-1]
targetY = loan_data.iloc[:,-1]


input_array = input_data.values

rescaled_input = MinMaxScaler(feature_range = (0,1)).fit_transform(input_array)

normalized_input = Normalizer().fit(rescaled_input).transform(rescaled_input)


normalized_inputX = pd.DataFrame(normalized_input, columns = input_data.columns)

normalized_inputX.describe()
```

```python
models = []
models.append(('LOR',LogisticRegression()))
models.append(('DTC', DecisionTreeClassifier()))
models.append(('KNC',KNeighborsClassifier()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('GNB',GaussianNB()))
models.append(('SVC',SVC()))

for name,model in models:
    sfs_get = sfs(model,k_features= 7, forward = True, verbose = 2, scoring='neg_mean_squared_error')
    sfs_get = sfs_get.fit(normalized_inputX,targetY)
    features_selected = list(sfs_get.k_feature_names_)
    print(f'{name} - {features_selected}')
```

```python
X_train, y_train = SMOTE().fit_resample(X_train, y_train)
X_test, y_test = SMOTE().fit_resample(X_test, y_test)
X_train, y_train = SMOTE().fit_resample(X_train, y_train)
X_test, y_test = SMOTE().fit_resample(X_test, y_test)
print("X_train shape", X_train.shape)
print("X_test shape", X_test.shape)
print("y_train shape", y_train.shape)
print("y_test shape", y_test.shape)
```

```python
models = []
models.append(('LOR',LogisticRegression()))
models.append(('DTC', DecisionTreeClassifier()))
models.append(('KNC',KNeighborsClassifier()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('GNB',GaussianNB()))
models.append(('SVC',SVC()))
```

```python
results = []
names = []
scoring = 'accuracy'
numFolds = 10 
for name, model in models:
    kfold = KFold(n_splits = numFolds, random_state = 50, shuffle = True)#, random_state = 7
    cv_results =cross_val_score(model, normalized_inputX, targetY, cv=kfold, scoring = scoring)
    predict = cross_val_predict(model,normalized_inputX, targetY, cv= kfold)
    results.append(cv_results)
    names.append(name)
    msg = (f'{name}, accuracy - {cv_results.mean()}, error - {cv_results.std()} - {make_scorer(accuracy_score)}') #cv_reslts.mean()-accuracyn score and cv_results.std -error
    print(msg)
    
```

```python

for name, model in models:
    model.fit(normalized_inputX,targetY)
    predictions = model.predict(X_test)
    print(f'{name}')    
    msg = (f'accuracy - {accuracy_score(y_test, predictions)}, F1-Score - {f1_score(y_test, predictions)}') #cv_reslts.mean()-accuracyn score and cv_results.std -error
    print(msg)
    print(f'Confusion matrix{confusion_matrix(y_test, predictions)}')
    print('')
    print('')
```
